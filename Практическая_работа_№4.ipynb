```python
class Transport:
  def __init__(self, n, s):
    self.name = n
    self.speed = s
  def travel(self):
    return "по миру"
  def drive(self):
    return "человеком"
  def carry(self):
    return
  def __str__(self):
    return("Название - {} \n Средняя скорость - {} \n Путешествует {} \n Управляется {} \n".format(self.name, self.speed, self.travel(), self.drive(), self.carry()))
class Car(Transport):
  def __init__(self, n, s, m):
    Transport.__init__(self, n, s)
    self.manufacture = m
  def buy(self):
    return "2500000"
  def sell(self):
    return "2000000"
  def __str__(self):
    return("Название - {} \n Скорость - {} \n Производитель - {} \n  Была куплена за  {} \n Продается за {} \n".format(self.name, self.speed, self.manufacture, self.buy(), self.sell()))
class Train(Transport):
  def __init__(self, n, s):
    Transport.__init__(self, n, s)
  def deliver(self):
    return "пассажиров"
  def can(self):
    return "опоздать на перрон"
  def __str__(self):
    return("Название - {} \n Скорость - {} \n Перевозит {} \n Может {} \n ".format(self.name, self.speed, self.deliver(), self.can()))
class Express(Train):
  def __init__(self, n, s, r):
    Transport.__init__(self, n, s)
    self.rout = r
  def teleport(self):
    return "в Хогвартс"
  def can(self):
    return "предлжить вкусняшки"
  def __str__(self):
    return("Название - {} \n Скорость - {} \n Направление - {} \n Телепортирует {} \n Может {} \n".format(self.name, self.speed, self.rout, self.teleport(), self.can()))
TR = Transport("Наземный транспорт", "65")
print(TR)
c1 = Car("Мазда", "150", "Япония")
print(c1)
t1 = Train("Сапсан", "250")
print(t1)
e1 = Express("Хогвартс Экспресс", "300", "Мир магии и волшебства")
print(e1)
```
